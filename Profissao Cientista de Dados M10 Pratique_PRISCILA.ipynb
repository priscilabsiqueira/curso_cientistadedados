```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv("ecommerce_preparados.csv")

df.head(10)
```

```python
# GRÁFICO DE HISTOGRAMA

plt.figure(figsize=(8, 6))
sns.histplot(df['N_Avaliações'].dropna(), bins=20, kde=False, edgecolor='black')
plt.title('Distribuição de Número de avaliações')
plt.xlabel('Número de Avaliações')
plt.ylabel('Frequência')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

# A maioria dos itens não recebeu muitas avaliações por algum motivo.
Apenas alguns itens conseguiram acumular um grande número de avaliações.

```python
# GRÁFICO DE DISPERSÃO

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Desconto_MinMax', y='Qtd_Vendidos_Cod', data=df)
plt.title('Dispersão entre Desconto_MinMax e Qtd_Vendidos_Cod')
plt.xlabel('Desconto_MinMax')
plt.ylabel('Qtd_Vendidos_Cod')
plt.grid(True)
plt.show()

```

# O gráfico mostra que não existe uma relação direta entre desconto e quantidade vendidos.

```python
# MAPA DE CALOR

# Seleciona apenas as colunas numéricas

df_numericas = df.select_dtypes(include=['float64', 'int64'])

# Calcula a matriz de correlação (com pairwise completo para lidar com NaNs)

correlacao = df_numericas.corr()

# Cria o mapa de calor

plt.figure(figsize=(10, 8))

sns.heatmap(correlacao, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)

plt.title('Mapa de Calor das Correlações')

plt.tight_layout()

plt.show()
```

# Pontos vermelhos indicam fortes relações entre as duas variáveis

```python
# GRÁFICO DE BARRAS

print(df['Qtd_Vendidos_Cod'].isnull().sum())
print(df['Marca_Cod'].isnull().sum())

import matplotlib.pyplot as plt
df['Marca_Cod'].hist()

print(df['Marca_Cod'].value_counts())
Qtd_Vendidos_Marca = df.groupby('Qtd_Vendidos_Cod')['Marca_Cod'].mean()
plt.bar(Qtd_Vendidos_Marca.index, Qtd_Vendidos_Marca.values, color='blue')

plt.title('Média de quantidade vendidos', fontsize=16)
plt.xlabel('Qtd_Vendidos_Cod', fontsize=12)
plt.ylabel('Marca_Cod', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

# Pode indicar qual marca foi mais ou menos vendida (dependendo do significado exato dos números em Marca_cod

```python
# Gráfico de pizza

# frequência de cada gênero

contagem_generos = df['Gênero'].value_counts()
labels = contagem_generos.index
sizes = contagem_generos.values

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.show()
```

# A maior frequência de gênero é o feminino, seguido pelo masculino.

```python
# Gráfico de densidade

df['Nota'].plot.density(figsize=(8, 6))

plt.title('Gráfico de Densidade das Notas dos Produtos')
plt.xlabel('Nota')
plt.ylabel('Densidade', fontsize=12)
plt.grid(True)
plt.show()

```

# A maior concentração de produtos avaliados recebeu avaliações entre 4 e 5.
