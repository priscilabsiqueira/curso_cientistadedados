```python
import pandas as pd

data = {
    'Nome': ['Alice', 'Joao', 'Charlie', 'David', 'Eva', 'Diego', 'Denize', 'Claudio'],
    'Idade': [25, 30, 35, 40, 45, 60, 22, 24],
    'Profissão': ['Engenheiro', 'Médico', 'Professor', 'Advogado', 'Médico','Engenheiro', 'Estudante','Estudante'],
    'Salário': ['4500', '8000', '5000', '10000', '12000','15000', '1200','1500'],
    'Limite_Credito': ['2500', '4000', '4000', '1000', '10000','2000', '500','250'],
    'Historico_Inadimplencia': ['0', '0', '0', '1', '0','1', '0','1'],
    'Estado_Civil': ['Casamento', 'Casamento', 'Solteiro', 'Solteiro', 'Casamento','Solteiro', 'Solteiro','Solteiro'],
    'Imovel_Proprio': ['0', '0', '0', '1', '1','1', '0','0']
}

df = pd.DataFrame(data)
df.head(5)

```

```python
import matplotlib.pyplot as plt

# 1 - GRÁFICO DE BARRAS

# Figura de imagem a ser gerada, escolhendo o tamanho:

plt.figure(figsize=(10, 6))

# Converter Limite em números e Média de crédito por profissão:

df['Limite_Credito'] = pd.to_numeric(df['Limite_Credito'], errors='coerce')
limite_medio_por_profissao = df.groupby('Profissão')['Limite_Credito'].mean()

for i, valor in enumerate(limite_medio_por_profissao):
    plt.text(i, valor, f'R${valor:.2f}' , ha='center', va='bottom', fontsize=10)

# Indicamos aqui os índices do DataFrame no eixo x e os salários no eixo y:

plt.bar(limite_medio_por_profissao.index, limite_medio_por_profissao.values, color='pink')

# Restante da personalização do gráfico:

plt.title('Limite Médio de Crédito por Profissão')
plt.xlabel('Profissão')
plt.ylabel('Limite Médio de Crédito (R$)')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

```
